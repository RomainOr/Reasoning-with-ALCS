```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_cartpole_beacs_debiased = pd.read_csv('df_cartpole_beacs_debiased.csv')
df_cartpole_beacs_biased = pd.read_csv('df_cartpole_beacs_biased.csv')
df_cartpole_acs2_debiased = pd.read_csv('df_cartpole_acs2_debiased.csv')
```

```python
palette = sns.color_palette('pastel')
sns.lineplot(x='Episode', y='value', data=pd.melt(df_cartpole_beacs_debiased, ['Episode']), color=palette[0], label="BEACS-debiased")
sns.lineplot(x='Episode', y='value', data=pd.melt(df_cartpole_beacs_biased, ['Episode']), color=palette[1], label="BEACS")
sns.lineplot(x='Episode', y='value', data=pd.melt(df_cartpole_acs2_debiased, ['Episode']), color=palette[2], label="ACS2-debiased")
plt.xlabel("Episode", fontsize = 14) # label for x-axis
plt.ylabel("Average score", fontsize = 14) # label for y-axis
plt.legend()
plt.show()
```
