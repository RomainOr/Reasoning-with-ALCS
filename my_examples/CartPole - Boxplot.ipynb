```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_cartpole_beacs_debiased = pd.read_csv('df_cartpole_beacs_debiased.csv')
df_cartpole_beacs = pd.read_csv('df_cartpole_beacs.csv')
df_cartpole_acs2_debiased = pd.read_csv('df_cartpole_acs2_debiased.csv')
df_cartpole_acs2 = pd.read_csv('df_cartpole_acs2.csv')
```

```python
def subplot_cartpole_performance(ax, df_data, color_palette, label):
    sns.lineplot(ax=ax, x='Episode', y='value', data=pd.melt(df_data, ['Episode']), color=color_palette, legend=False)
    ax.axhline(y = 475.0, color = 'red', linestyle = 'dashed')
    ax.text(2, 477, 'Solving threshold', color='red')
    ax.set_yticks(yticks, labels=yticklabels)
    ax.axvline(x = 500, color = 'gray', linestyle = 'dotted')
    ax.text(400, 250, 'Exploration', color='gray', ha='center', va='center', rotation='vertical')
    ax.text(600, 250, 'Exploitation', color='gray', ha='center', va='center', rotation='vertical')
    ax.set_xticks(xticks, labels=xticklabels)
    ax.set_xlabel("Episode")
    ax.set_ylabel("Average episode length")
    ax.set_title(label)
    ax.grid()

palette = sns.color_palette()
yticks = [0, 100, 200, 300, 400, 475, 500]
yticklabels = [0, 100, 200, 300, 400, 475, 500]
xticks = [0, 200, 400, 600, 800, 1000]
xticklabels = [0, 200, 400, 600, 800, 1000]

plt.figure(dpi=300)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4,sharex=True, sharey=True, figsize=(20, 5))
subplot_cartpole_performance(ax1, df_cartpole_beacs_debiased, palette[1], "d-BEACS")
subplot_cartpole_performance(ax2, df_cartpole_beacs, palette[0], "BEACS")
ax2.yaxis.set_tick_params(labelleft=True, labelcolor="black")
subplot_cartpole_performance(ax3, df_cartpole_acs2_debiased, palette[4], "d-ACS2")
ax3.yaxis.set_tick_params(labelleft=True, labelcolor="black")
subplot_cartpole_performance(ax4, df_cartpole_acs2, palette[2], "ACS2")
ax4.yaxis.set_tick_params(labelleft=True, labelcolor="black")

plt.show()
```
